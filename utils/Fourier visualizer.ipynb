```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
SAMPLING_RATE = 1000  # Hz (Must be > 2*max frequency, Nyquist Theorem)
DURATION = 1.0        # Seconds
NUM_SAMPLES = int(SAMPLING_RATE * DURATION)

# Time array
t = np.linspace(0, DURATION, NUM_SAMPLES, endpoint=False)

# Signal: 2*sin(2π*3*t) + 4*sin(2π*5*t)
signal = 2 * np.sin(2 * np.pi * 3 * t) + 4 * np.sin(2 * np.pi * 10 * t)

# Compute FFT
fft_result = np.fft.fft(signal)
fft_magnitude = np.abs(fft_result)  # Magnitude spectrum
frequencies = np.fft.fftfreq(NUM_SAMPLES, 1/SAMPLING_RATE)  # Frequency bins

# Plotting
plt.figure(figsize=(12, 6))

# Time Domain
plt.subplot(1, 2, 1)
plt.plot(t, signal)
plt.title("Time Domain Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid()

# Frequency Domain
positive_freq_mask = frequencies >= 0
plt.subplot(1, 2, 2)
plt.plot(frequencies[positive_freq_mask], fft_magnitude[positive_freq_mask])
plt.title("Frequency Domain (FFT)")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.grid()

plt.tight_layout()
plt.show()
```
